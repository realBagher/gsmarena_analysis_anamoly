```python
import json
import pandas as pd 
```

```python
def convert_list_to_dict(obj):
    if isinstance(obj, dict):
        for key, value in obj.items():
            obj[key] = convert_list_to_dict(value)
    elif isinstance(obj, list) and all(isinstance(item, dict) for item in obj):
        combined_dict = {}
        for item in obj:
            combined_dict.update(item)
        return combined_dict
    return obj
```

```python
files = ['Alcatel_data.json','Apple_data.json' , 'Asus_data.json' , 'Blu_data.json' , 'HTC_data.json' , 'Huawei_data.json' , 'Infinix_data.json' , 'Lenovo_data.json' ,
         'LG_data.json' , 'Nokia_data.json' , 'Samsung_data.json' , 'Sony_data.json' , 'Xiaomi_data.json' , 'ZTE_data.json']
```

```python
def Screan_Ratio_df(json_file) :
    file = open(json_file)
    data = json.load(file)

    brand = list(data.keys())[0]

    device_list = list(data[brand].keys())
    
    screen_df = pd.DataFrame(index = device_list , columns = ['screen_to_body_ratio'] ) 
    for device in device_list[:] :
        try : 
            d = convert_list_to_dict(data[brand][device]['Display'])
            #print(data[brand][device]['Display']) 
            #print(float(d['Size'].split(',')[1].split('(')[1].split(' ')[0][1:-1]))
            screen_df.loc[device]['screen_to_body_ratio'] = float(d['Size'].split(',')[1].split('(')[1].split(' ')[0][1:-1])
        except :
            screen_df.loc[device]['screen_to_body_ratio'] = None
    
    return screen_df
```

```python
Screan_Ratio = pd.DataFrame(columns= ['screen_to_body_ratio'])
for file in files :
    new_df = Screan_Ratio_df(file)
    Screan_Ratio = pd.concat([Screan_Ratio , new_df] , axis=0)
```

```python
Screan_Ratio.reset_index(inplace=True)
```

```python
Screan_Ratio.rename(columns={'index' : 'DeviceName'} , inplace=True)
```
