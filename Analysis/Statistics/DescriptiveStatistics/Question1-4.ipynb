```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## : دیتا را میخوانیم

```python
dnet = pd.read_csv('E:/quera tamrin/mini p/Network.csv')
```

## .چک میکنیم که دیتا را درست خوانده باشیم

```python
dnet.head()
```

## .گروه بندی میکنیم و تعداد هرکدام را محاسبه میکنیم network داده ها را بر اساس ستون 

## .سپس نمودار های آن را رسم میکنیم

* سوال 1 آمار توصیفی


```python
network_counts = dnet['Network'].value_counts()
#plotting
plt.figure(figsize=(8, 6))
network_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Mobile Devices Among Different Network Technologies')
plt.xlabel('Network Type')
plt.ylabel('Number of Devices')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

## سوال 2

##  : داده های فایل دستگاه ها را میخوانیم

## .سپس ماتریس همبستگی که سوال خواسته بود را رسم میکنیم







```python
#2
ddev = pd.read_csv('E:/quera tamrin/mini p/Devices.csv') 

ddev['TotalPixels'] = ddev['PixelX'] * ddev['PixelY']
selected_columns = ['DisplaySize', 'DisplayArea', 'Weight', 'TotalPixels', 'ScreenToBodyRatio', 'BatterymAh', 'PPIDensity']
ddev_selected = ddev[selected_columns]
correlation_matrix = ddev_selected.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

## در سوال 4 از همان دیتایی که مختص محصولات بود و قبلا خوانده بودیم استفاده میکنیم

```python
#4
android_counts = ddev['OS'].value_counts()

# Select the top ten versions
top_ten_android = android_counts.head(10)

# Plot the top ten versions using a bar plot
plt.figure(figsize=(10, 6))
top_ten_android.plot(kind='bar', color='skyblue')
plt.title('Top Ten Versions of Android in Terms of Repeated Use')
plt.xlabel('Android Version')
plt.ylabel('Number of Devices')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

## امتیازی
## رابطه ی بین وزن کالا و سایز

```python
# Convert 'DisplaySize' and 'Weight' columns to numeric data types
ddev['DisplaySize'] = pd.to_numeric(ddev['DisplaySize'], errors='coerce')
ddev.dropna(inplace=True)

# Plotting 
plt.figure(figsize=(10, 6))
plt.scatter(ddev['DisplaySize'], ddev['Weight'], color='blue')
plt.title('Scatter plot of DisplaySize vs Weight')
plt.xlabel('DisplaySize')
plt.ylabel('Weight')
plt.grid(True)
plt.show()

# Calculate the correlation coefficient
correlation_coefficient = ddev['DisplaySize'].corr(ddev['Weight'])
print("Correlation Coefficient between DisplaySize and Weight:", correlation_coefficient)


```
